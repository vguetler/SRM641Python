## Matplotlib

Matplotlib is a desktop plotting package designed for creating plots and figures suitable for publication. Matplotlib has spawned a number of add-on toolkits for data visualization that use matplotlib for their underlying plotting such as `seaborn`. `Pandas` uses `matplotlib` to create visualizations. Therefore, before we learn how to plot with pandas or seaborn, it's important to understand how matplotlib works.

For more information on matplotlib see: https://matplotlib.org/stable/users/explain/quick_start.html

## Plot Components

### Figure
Top-level object that holds the other plot components

Matplotlib graphs your data on `Figures`, each of which can contain one or more `Axes`, an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.). 

```python
fig = plt.figure()
```

### Axes
Individual plots contained within the Figure.

### Creating subplots
Simply specify the number of rows and columns to create:

```python
fig, axes = plt.subplots(1, 2)
```

```python
# Create three figures in a 2x2 format
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

### Creating Plot Layouts with gridspec

We can create subplots with varying sizes as well:

```python
fig = plt.figure(figsize=(8, 8))
gs = fig.add_gridspec(3, 3)
top_left = fig.add_subplot(gs[0, 0])
mid_left = fig.add_subplot(gs[1, 0])
top_right = fig.add_subplot(gs[:2, 1:])
bottom = fig.add_subplot(gs[2,:])
```

## Scatter plots

We can pass in a string specifying the style of the plot. This is of the form `[marker][linestyle][color]`. For example, we can make a black dashed line with '--k' or a red scatter plot with 'or':

```python
# Let's plot the high and low variables

plt.plot('high', 'low', 'or', data= df_fb.head(20))
```

```python
# Make a black dashed line with --k and a style 

plt.style.use('_mpl-gallery')
plt.plot('high', 'low', '--k', data= df_fb.head(20))
```

For more info about the string specifying style of the plot, see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

## Histograms

We will use the earthquakes data

```python
# plot a histogram of the variable magtype (magnitude type)
plt.hist(df_earthquakes.query('magType == "ml"').mag)
```

Bin size matters: the magnitude tend to be small and the distribution appears to be normal. However, note that bin size matters. Some cases we can change the number of bins the data is divided into. 

If we make 2 histograms for this data using different number of bins, the distribution will look different:

```python
x = df_earthquakes.query('magType == "ml"').mag
fig, axes = plt.subplots(1, 2, figsize=(10, 3))
for ax, bins in zip(axes, [7, 35]):
    ax.hist(x, bins=bins)
    ax.set_title(f'bins param: {bins}')
```

### Saving plots

You can save the active figure to file using the figure object’s savefig instance method. The file type is inferred from the file extension. So if you used .pdf instead, you would get a PDF.

Use `plt.savefig()` to save the last created plot. To save a specific Figure object, use its `savefig()` method. One important option for publishing graphics is dpi, which controls the dots-per-inch resolution. To get the same plot as a PNG at 400 DPI, you would do:

```python
plt.savefig('plot.png', dpi=400)
```

### Cleaning up

It's important to close resources when we are done with them. We use `plt.close()` to do so. If we pass in nothing, it will close the last plot, but we can pass in the specific Figure object to close or say 'all' to close all Figure objects that are open. Let's close all the Figure objects that are open with plt.close():

```python
plt.close('all')
```
